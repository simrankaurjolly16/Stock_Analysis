# Corelation

```python
yahoo_data['ticker']='yahoo'
amc_data['ticker']='AMC'
azo_data['ticker']= 'AZO'
bam_data['ticker']= 'BAM'
fis_data['ticker']='FIS'
gmz_data['ticker']='GMZ'
gnaf_data['ticker']='GNAF'
gnk_data['ticker']='GNK'
gne_data['ticker']='GNE'
```

```python
stocks_list =[yahoo_data,amc_data,azo_data,bam_data,fis_data,gmz_data,gnaf_data,gnk_data,gne_data]
Stocks=pd.concat(stocks_list,sort='ticker')

#description de la BD cryptoStocks
Stocks.info()
print(Stocks.head())
print(Stocks.tail())
```

```python
#restreindre aux colonnes 'ticker', 'Adj Close' et 'Date'
Stocks_restraint= Stocks[['ticker','Adj Close']]

#voir les 5 dernières ligne de la BD
print(Stocks_restraint.head())
print(Stocks_restraint.tail())
```

```python
#indexer les colonnes Date et ticker pour calculer la variation sur la colonne Adj Close seule
Stocks_index=Stocks_restraint.set_index(['ticker'])
print(Stocks_index.head())
print(Stocks_index.tail())
#compter les données qui sont vides 
print(" non renseigné sur la colonne Adj Close ???")
Stocks_index.isnull().sum()
```

```python
Stocks_restraint= Stocks_restraint.reset_index()
```

```python
Stocks_restraint.head()
```

```python
Stocks_graph = Stocks_restraint.pivot(index='Date', columns='ticker',values='Adj Close' ) #index='Date', columns='ticker', values='Adj Close'
Stocks_graph.head()
```

```python
Stocks_fill=Stocks_graph.bfill().ffill()
Stocks_fill.head()
```

```python
ax= Stocks_fill.plot(kind='line', figsize=(15,10))
plt.legend(fontsize=16) 
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('prix', fontsize=16)
plt.show()
```

```python
#importer la bibliothèque seaborn
import seaborn as sns

#créer une matrice de corrélation de type Spearman (rang)
Stocks_corr = Stocks_fill.corr(method='spearman')


# représenter la matrice en graphique type 'heatmap'
graph = sns.clustermap(Stocks_corr, annot=True, annot_kws={"size": 10}, linewidths=.4, figsize=(15, 10))
plt.setp(graph.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.setp(graph.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
graph.fig.suptitle('Spearman')
plt.show()
print(Stocks_corr)


#créer une matrice de corrélation de type Pearson (rang)
Stocks_corr = Stocks_graph.corr(method='pearson')


# représenter la matrice en graphique type 'heatmap'
graph = sns.clustermap(Stocks_corr, annot=True, annot_kws={"size": 10}, linewidths=.4, figsize=(15, 10))
plt.setp(graph.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.setp(graph.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
graph.fig.suptitle('Pearson')
plt.show()
print(Stocks_corr)
```

```python
import seaborn
mask = np.zeros_like(Stocks_corr)
mask[np.triu_indices_from(mask)] = True
 #generate plot
seaborn.heatmap(Stocks_corr, cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2.5)
plt.yticks(rotation=0) 
plt.xticks(rotation=90) 
plt.show()
```
